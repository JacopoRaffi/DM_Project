## Plots on 'birth_year' data

```python
# Plot histogram of 'birth_year' data
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
# 10 bins
sb.histplot(dataset, x='birth_year', bins=10, ax=axes[0])
axes[0].set_title('10 Bins')
# 25 bins
sb.histplot(dataset, x='birth_year', bins=25, ax=axes[1])
axes[1].set_title('25 Bins')
#50 bins
sb.histplot(dataset, x='birth_year', bins=50, ax=axes[2])
axes[2].set_title('50 Bins')

plt.tight_layout() # Adjust the subplots to fit into the figure area.
plt.show()
```

## Plot on 'nationality' data

```python
# Count the occurrences
nationality_counts = dataset['nationality'].value_counts().reset_index()
nationality_counts.columns = ['nationality', 'count']

plt.figure(figsize=(20, 6))
sb.barplot(x='nationality', y='count', data=nationality_counts)

plt.xticks(rotation=90) # Rotate the x-axis labels
plt.title('Occurrences for each nation')
plt.show()
```

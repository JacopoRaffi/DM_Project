In this notebook, we perform a visual exploration to understand the structure of the attributes and identify potential outliers.

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
dataset = pd.read_csv('../data/complete_dataset_fe.csv', dtype={'cyclist_team': str})
dataset['date'] = pd.to_datetime(dataset['date'], format='%Y-%m-%d')
```

```python
dataset.info()
```

## Cyclist Outlier

```python
plt.figure(figsize=(10, 5))
sb.boxenplot(x='cyclist_age', data=dataset.drop_duplicates(subset=['cyclist_age', 'cyclist_url']))
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.xlabel('Cyclist Age', fontsize=15)
plt.show()
```

With regard to the attribute `cyclist_age`, the box plot display makes it possible to identify values roughly below 20 and values above 40 as outliers.

## Races Outliers

```python
tmp_df = tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'year': dataset['date'].dt.year
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='year', data=tmp_df)
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.xlabel('Year', fontsize=15)
plt.show()
```

The boxenplot displayed above shows the quantile levels distributions of the years in which the races in our dataset took place. No clear outliers are visible from a qualitative analysis.

```python
plt.figure(figsize=(15, 5))
sb.boxenplot(x='length', data=dataset.drop_duplicates(subset=['race_url', 'date']))
plt.gca().xaxis.set_major_locator(MultipleLocator(20))
plt.xlabel('Length', fontsize=15)
plt.show()
```

With regard to the races attribute `length`, the box plot display makes it possible to identify values roughly below 7 Km and values above 290 Km as outliers.

```python
plt.figure(figsize=(15, 5))
sb.boxenplot(x='climb_total', data=dataset.drop_duplicates(subset=['race_url', 'date']))
plt.gca().xaxis.set_major_locator(MultipleLocator(0.5))
plt.xlabel('Total Climb', fontsize=15)
plt.show()
```

With regard to the races attribute "climb_total", the box plot display makes it possible to identify values roughly below 0.1 Km and values above 6 Km as outliers.

```python
tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'race_url': dataset['race_url'],
    'year': dataset['date'].dt.year,
    'position': dataset['position']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='position', data=tmp_df)
plt.gca().xaxis.set_major_locator(MultipleLocator(0.1))
plt.xlabel('Position', fontsize=15)
plt.show()
```

The boxenplot displayed above shows the quantile levels distributions of the positions of the cyclists in the races registered in our dataset. No clear outliers are visible from a qualitative analysis. This is an expected result since our normalization process "destroyed" the absolute values and distributions of the starting `position` attribute.

```python
tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'race_url': dataset['race_url'],
    'year': dataset['date'].dt.year,
    'delta': dataset['delta']
}).drop_duplicates().dropna()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='delta', data=tmp_df, log_scale=False)
plt.gca().xaxis.set_major_locator(MultipleLocator(5000))
plt.xlabel('Delta', fontsize=15)
plt.show()
```

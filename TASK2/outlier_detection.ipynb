```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
cyclist = pd.read_csv('../data/cyclists.csv')
races = pd.read_csv('../data/races.csv')

#TODO: remove it when you use cleaned data
#TODO: change the attribute name when you use cleaned data
races['date'] = races['date'].str.split(' ').str[0]
races['date'] = pd.to_datetime(races['date'], format='%Y-%m-%d')
```

## Races Outliers

```python
tmp_df = tmp_df = pd.DataFrame({
    'race_name': races['name'],
    'year': races['date'].dt.year
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxplot(x='year', data=tmp_df)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sb.boxplot(x='length', data=races.drop_duplicates(subset=['_url', 'date']))
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sb.boxplot(x='climb_total', data=races.drop_duplicates(subset=['_url', 'date']))
plt.show()
```

```python
tmp_df = tmp_df = pd.DataFrame({
    'race_name': races['name'],
    'year': races['date'].dt.year,
    'startlist_quality': races['startlist_quality']
}).drop_duplicates()

plt.figure(figsize=(15, 5))
sb.boxplot(x='startlist_quality', data=races.drop_duplicates(subset=['_url', 'date']))
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': races['name'],
    'race_url': races['_url'],
    'year': races['date'].dt.year,
    'points': races['points']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxplot(x='points', data=tmp_df)
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': races['name'],
    'race_url': races['_url'],
    'year': races['date'].dt.year,
    'uci_points': races['uci_points']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxplot(x='uci_points', data=tmp_df)
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': races['name'],
    'race_url': races['_url'],
    'year': races['date'].dt.year,
    'position': races['position']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxplot(x='position', data=tmp_df)
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': races['name'],
    'race_url': races['_url'],
    'year': races['date'].dt.year,
    'delta': races['delta']
}).drop_duplicates().dropna()

plt.figure(figsize=(10, 5))
sb.boxplot(x='delta', data=tmp_df, log_scale=True) #TODO: capire se  giusto fare scala logaritmica
plt.show()
```

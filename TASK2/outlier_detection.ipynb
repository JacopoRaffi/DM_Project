```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
dataset = pd.read_csv('../data/complete_dataset_fe.csv', dtype={'cyclist_team': str})

#TODO: remove it when you use cleaned data
#TODO: change the attribute name when you use cleaned data
dataset['date'] = pd.to_datetime(dataset['date'], format='%Y-%m-%d')
```

```python
dataset.info()
```

## Cyclist Outlier

```python
plt.figure(figsize=(10, 5))
sb.boxenplot(x='birth_year', data=dataset.drop_duplicates(subset=['birth_year', 'cyclist_url']))
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.boxenplot(x='weight', data=dataset.drop_duplicates(subset=['weight', 'cyclist_url']))
plt.gca().xaxis.set_major_locator(MultipleLocator(3))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sb.boxenplot(x='height', data=dataset.drop_duplicates(subset=['height', 'cyclist_url']))
plt.gca().xaxis.set_major_locator(MultipleLocator(3))
plt.show()
```

## Races Outliers

```python
tmp_df = tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'year': dataset['date'].dt.year
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='year', data=tmp_df)
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sb.boxenplot(x='length', data=dataset.drop_duplicates(subset=['race_url', 'date']))
plt.gca().xaxis.set_major_locator(MultipleLocator(20))
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sb.boxenplot(x='climb_total', data=dataset.drop_duplicates(subset=['race_url', 'date']))
plt.gca().xaxis.set_major_locator(MultipleLocator(0.5))
plt.show()
```

```python
tmp_df = tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'year': dataset['date'].dt.year,
    'startlist_quality': dataset['startlist_quality']
}).drop_duplicates()

plt.figure(figsize=(15, 5))
sb.boxenplot(x='startlist_quality', data=dataset.drop_duplicates(subset=['race_url', 'date']))
plt.gca().xaxis.set_major_locator(MultipleLocator(100))
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'race_url': dataset['race_url'],
    'year': dataset['date'].dt.year,
    'points': dataset['points']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='points', data=tmp_df)
plt.gca().xaxis.set_major_locator(MultipleLocator(20))
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'race_url': dataset['race_url'],
    'year': dataset['date'].dt.year,
    'uci_points': dataset['uci_points']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='uci_points', data=tmp_df)
plt.gca().xaxis.set_major_locator(MultipleLocator(50))
plt.show()
```

```python
tmp_df = pd.DataFrame({
    'race_name': dataset['race_name'],
    'race_url': dataset['race_url'],
    'year': dataset['date'].dt.year,
    'position': dataset['position']
}).drop_duplicates()

plt.figure(figsize=(10, 5))
sb.boxenplot(x='position', data=tmp_df)
plt.gca().xaxis.set_major_locator(MultipleLocator(0.1))
plt.show()
```
